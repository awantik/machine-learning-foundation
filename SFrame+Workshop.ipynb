```python
csv_insurance = SFrame('D:\Data\earthquake-database\database.csv')
```

```python
csv_insurance.head()
```

```python
earthquake_results = csv_insurance['Date','Time','Latitude','Longitude','Type','Depth','Magnitude']
```

```python
earthquake_results.head()
```

```python
def getScale(row):
    if row['Magnitude'] > 6:
        return "High"
    else:
        return "Medium"

earthquake_results['result'] = earthquake_results.apply(getScale)
```

```python
earthquake_results.head()
```

```python
summary = earthquake_results[ earthquake_results['result'] == "High"]['Date','Latitude','Longitude']
```

```python
summary.export_csv('D:\Data\File.csv')
```

```python
summary.export_json('D:\Data\File.json')
```

```python
sf = graphlab.SFrame({'letter': ['a', 'b', 'c'], 'number': [1, 2, 3]})
```

```python
sf.head()
```

```python
def gen_rows(x):
    print x
    a = [x['number']*2,x['letter']]
    b = [x['number']*3,x['letter']]
    return [a,b]

sf.flat_map(['number', 'letter'],gen_rows)
```
