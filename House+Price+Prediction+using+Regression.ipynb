```python
train_data, test_data = home.random_split(0.8, seed = 0)
```

```python
 help(graphlab.linear_regression)
```

```python
sqft_model = graphlab.linear_regression.create(train_data, target ='price', features = ['sqft_living'])
```

```python
sqft_model.evaluate(test_data)
```

```python
more_features_model = graphlab.linear_regression.create(train_data, target ='price', features = ['sqft_living','bedrooms','bathrooms','zipcode', 'floors'])
```

```python
more_features_model.evaluate(test_data)
```
