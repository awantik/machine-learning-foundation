```python
comp = gl.SFrame('D:\Data\company_master_data_upto_Mar_2015_Karnataka.csv')
```

```python
comp = comp.dropna()
```

```python
# Data cleaning, removal of comma
def fun(x):
    return float((x['AUTHORIZED_CAPITAL']).replace(',',''))
    
comp['AUTH_CAP'] = comp.apply(fun)
```

```python
comp.head()
```

```python
comp.sort('AUTH_CAP', ascending=True)['COMPANY_NAME','AUTHORIZED_CAPITAL']
```

```python
def get_year(x):
    yr = x['DATE_OF_REGISTRATION']
    return yr.split('-')[2]
    
comp['YEAR'] = comp.apply(get_year)
```

```python
comp['YEAR'] = comp.apply(lambda x: int(x['YEAR']))
```

```python
import graphlab.aggregate as agg
comp.groupby(key_columns='YEAR', operations= agg.COUNT())
```
